# 1. Convert 'Age' column to a NumPy array. 

```python
numpy = df['Age'].to_numpy()
numpy
```

# 2. Find the average age of players. 

```python
x=np.array(df["Age"].mean())
print(x)
```

# 3. Find the oldest and youngest player age. 

```python
max_age=np.array(df['Age'].max())
print(max_age)
min_age=np.array(df['Age'].min())
print(min_age)
```

# 4. Count the total number of players above 30 years old.

```python
count=np.array((df["Age"]>30).sum())
print(count)
```

# 5. Find the median batting average. 

```python
median = np.array(df["Batting Average"].median())
print(median)
```

# 6. Find the standard deviation of strike rates. 

```python
str=np.array(df['Strike Rate'].std())
print(str)
```

# 7. Find the 90th percentile of batting average.

```python
percentile_90 = np.percentile(df["Batting Average"],90)

print(percentile_90)
```

# 8. Find the number of players with strike rate above 150. 

```python
p = np.array(df["Strike Rate"] > 150).sum()
p
```

# 9. Reshape the first 20 batting averages into a 4x5 matrix. 

```python
reshaped_avg = np.array(df["Batting Average"][:20]).reshape(4,5)
print(reshaped_avg)
```

# 10. Compute row-wise sum of reshaped matrix. 

```python
row_sum=np.sum(reshaped_avg,axis=1)
print(row_sum)
```

# 11. Compute column-wise mean of reshaped matrix. 

```python
col_sum=np.sum(reshaped_avg,axis=0)
print(col_sum)
```

# 12. Transpose the reshaped matrix. 

```python
transpose=reshaped_avg.T
print(transpose)
```

# 13. Find the variance of batting average. 

```python
variance=np.array(df["Batting Average"].var())
print(variance)
```

# 14. Stack age and matches played horizontally. 

```python
a=np.array(df["Age"])
b=np.array(df["Matches Played"])
stacked=np.stack((a,b))
print(stacked)
```

# 15. Split the batting average array into 5 equal parts. 

```python
s=np.array(df["Batting Average"])
split_array=np.split(s,5)
print(split_array)
```
